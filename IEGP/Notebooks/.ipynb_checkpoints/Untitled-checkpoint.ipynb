```python
def phi(M,m,a,p):
    part1=0.4*np.log(10)
    paw1= -0.4*(M-m)*(a+1)
    paw2= -np.power(10,-0.4*(M-m))
    part2=np.power(10,paw1)
    part3=np.exp(paw2)
    
    phi_value = part1*part2*part3*p
    
    return phi_value
```

```python
popt,pcov = curve_fit(phi,bw_z4.M,bw_z4.phi, bounds=(-np.infty,np.infty))
pcov
```

```python
plt.figure(figsize=(15,10))
plt.errorbar(bw_z4.M,bw_z4.phi, bw_z4.err_phi,marker='o',ms=5, mfc='b',ecolor="b",linestyle="")
plt.plot(bw_z4.M, phi(bw_z4.M,popt[0],popt[1],popt[2]),'r-')
plt.yscale('log')
```

## Example PyMC3

```python
true_mu = 0.1
data = true_mu + np.random.randn(100)

with pm.Model() as model:
    mu = pm.Normal('mu', mu=0, sigma=1)
    y = pm.Normal('y', mu=mu, sigma=1, observed=data)
```

```python
with model:
    trace = pm.sample(5000, cores=4, nuts_kwargs={'target_accept': 0.95})
```

```python
pm.summary(trace)
```

#### Auto Ejemplo

```python
x = np.arange(0,4 * np.pi,4 * np.pi / 100)
y_true = np.sin(x)**2 + np.cos(3*x)
y = y_true + np.random.normal(0,.2,y_true.shape[0])

plt.plot(x,y_true,'r-')
plt.plot(x,y,'b*')
```

```python
with pm.Model() as model:
    b = pm.Uniform('b',0,10)
    
    mu = tt.pow(tt.sin(x), 2) + tt.cos(b*x)
    y_obs = pm.Normal('y_obs', mu=mu, sigma=.2, observed=y)
    trace = pm.sample(5000,chains=4,nuts_kwargs={'target_accept': 0.95})
    
pm.traceplot(trace)
pm.summary(trace)
```
