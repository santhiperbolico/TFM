```python
import numpy as np
from scipy.integrate import quad
import math
import matplotlib.pyplot as plt
```

** define values **

```python
h = 0.67
wb = 0.022
wm = 0.14
wr = 4.17e-5
ns = 0.97
H0 = h/2998 # en 1/Mpc
zdec = 1090.3
kD = 0.14 #1/Mpc
kpiv = 0.05
```

** define axis labels **

```python
xlab = 'multipole moment l'; ylab='$l(l+1)C_l/2\pi$ in $(\mu K)^2$'
titstr = 'angular spectrum of $(\Delta T/T)^2$'
```

** define functions **

```python
def E(z, hf, wmf, wrf):
    res = np.sqrt(wmf*pow(1+z, 3) + wrf*pow(1+z, 4) + pow(hf,2)-wrf-wmf)/hf
    return(res)

def R(z, wbf):
    return(3.04e4/(1+z)*wbf)

def dAc(z, hf, wmf, wrf):
    return( quad(lambda zz: 1/E(zz, hf, wmf, wrf), 0, z)[0]/(hf/2998) )

def cs(z, wbf):
    return( np.sqrt( (1/3)/(1+R(z, wbf)) ) ) 

def T(x):
    res = np.log(1+0.171*x)/(0.171*x) * pow(1 + 0.284*x + pow(1.18*x, 2) + pow(0.399*x, 3) + pow(0.490*x,4), -0.25)
    return(res)

def PR(k, nsf, As=2.2e-9, k0=kpiv):
    return( pow(k/k0, nsf-1)*As )

def rs(zdecf, wbf, hf, wmf, wrf):
    return( quad(lambda z: cs(z, wbf)/E(z, hf, wmf, wrf), zdecf, np.inf)[0]/(hf/2998) )

def bessel(l, x):
    if x > l: res = 1/(2*x*np.sqrt(pow(x,2) - pow(l,2)))
    else: res = 0
    return(res)
```

** implement $C_l$ **

```python
def C(l, 
      Rdec=None, rs_val = None, dAc_dec=None,
      kDf=kD, hf=h, wbf=wb, wmf=wm, nsf=ns, zdecf=zdec, wrf=wr
      ):
    if not Rdec: Rdec = R(zdecf, wbf)
    if not rs_val: rs_val = rs(zdecf, wbf, hf, wmf, wrf)
    if not dAc_dec: dAc_dec = dAc(zdecf, hf, wmf, wrf)
        
    keq = 0.073*wmf
    
    # function to integrate
    def intfunc(k):
        m1 = PR(k, nsf)/k
        m2 = pow( -Rdec * T(k/keq) + 5/3*np.cos(k*rs_val)*np.exp(-k**2/kD**2) ,2)
        m3 = bessel(l, k*dAc_dec)
        return(m1 * m2 * m3)
        
    res = 1e12*4*math.pi*9/25 * quad(lambda k: intfunc(k), l/dAc_dec, 4000/dAc_dec)[0]
    return(res)
    
# vectorize C:
vC = np.vectorize(C)
```

** Objective 1: Using values of the cosmological standard model **

$z_{dec}$, $R(z_{dec})$, $r_s$ and $d^c_A(z_{dec})$ for the values of the comological standard model:

```python
print("-- zdec ----- R(zdec) ---- rs -------- dAc(zdec) ----")
print("  {:.1f}     {:f}  {:f}   {:f} ".format(zdec, R(zdec, wb), rs(zdec, wb, h, wm, wr), dAc(zdec, h, wm, wr)) )
```

Spectrum of $C_l$ for $\Lambda CDM$ model:

```python
l = np.linspace(100,1500, 1401)
Cl_val = vC(l)
Cl_val_rescale = l*(l+1)*Cl_val/(2*math.pi)
```

```python
plt.plot(l, Cl_val_rescale)
plt.xlabel(xlab); plt.ylabel(ylab)
plt.title(titstr)
plt.savefig('stanval.pdf')
plt.show()
plt.close()
```

** Objective 2: vary $R(z_{dec})$, $r_s$ and $d^c_A$ **

```python
Rvar = np.array([0.2, 0.5, 1, 2])
Cl_val_rvar = np.empty([len(Rvar), len(l)])
for i in range(len(Rvar)): Cl_val_rvar[i] = np.array(vC(l, Rdec=Rvar[i]*R(zdec, wb)))
    
for i in range(len(Rvar)):
    plt.plot(l, l*(l+1)*Cl_val_rvar[i]/(2*math.pi), label=str(Rvar[i])+'$\, R(z_{dec})$')
plt.legend()
plt.xlabel(xlab); plt.ylabel(ylab)
plt.title(titstr+' for varying $R(z_{dec})$')
plt.savefig('Rvar.pdf'); plt.show(); plt.close()
```

Apparently increasing $R(z_{dec})$ increases the even peaks and diminshes the odd ones.

```python
rsvar = np.array([0.9, 1, 1.1])
Cl_val_rsvar = np.empty([len(rsvar), len(l)])
for i in range(len(rsvar)): Cl_val_rsvar[i] = np.array(vC(l, rs_val=rsvar[i]*rs(zdec, wb, h, wm, wr)))
    
for i in range(len(rsvar)):
    plt.plot(l, l*(l+1)*Cl_val_rsvar[i]/(2*math.pi), label=str(rsvar[i])+'$\, r_s$')
plt.legend()
plt.xlabel(xlab); plt.ylabel(ylab)
plt.title(titstr+' for varying $r_s$')
plt.savefig('rsvar.pdf'); plt.show(); plt.close()
```

Diminishing $r_s$ compresses the spectrum. As $r_s$ only appears inside the cosine this is only logical.

```python
dvar = np.array([0.9, 1, 1.1])
Cl_val_dvar = np.empty([len(dvar), len(l)])
for i in range(len(dvar)): 
    Cl_val_dvar[i] = np.array(vC(l, dAc_dec=dvar[i]*dAc(zdec, h, wm, wr)))
    
for i in range(len(dvar)):
    plt.plot(l, l*(l+1)*Cl_val_dvar[i]/(2*math.pi), label=str(dvar[i])+'$\, d^c_A$')
plt.legend()
plt.xlabel(xlab); plt.ylabel(ylab)
plt.title(titstr+' for varying $d^c_A$')
plt.savefig('dvar.pdf'); plt.show(); plt.close()
```

An increase in $d_A^c(z_{dec})$ does (almost) not change the amplitude of the peaks but rather their position because it only appears as the proportional constant between scale $k$ and the multipole moment $l$. In this approximation anyway it affects the amplitude a tiny bit because it is used to approximate the integration boundaries in $C_l$.

** Objective 3: vary $h$ **

```python
hvar = np.array([0.5, h, 0.8])
Cl_val_hvar = np.empty([len(hvar), len(l)])
for i in range(len(hvar)):
    Cl_val_hvar[i] = np.array(vC(l, hf=hvar[i]))
    
for i in range(len(hvar)):
    plt.plot(l, l*(l+1)*Cl_val_hvar[i]/(2*math.pi), label=str(hvar[i]))
plt.legend(); plt.xlabel(xlab); plt.ylabel(ylab)
plt.title(titstr+' for varying $h$')
plt.savefig('hvar.pdf'); plt.show(); plt.close()
```

Varying $h$ changes $E(z) = h^{-1}(\omega_m (1+z)^3 + \omega_r (1+z)^4 + \omega _{\Lambda})^{1/2}$. This could potentially effect 
$$ r_s = H_0^{-1} \int _{z_{dec}} ^{\infty} \frac{c_s(z)}{E(z)}dz $$
and
$$ d_A^c(z_{dec}) = H_0 ^{-1}  \int _0 ^{z_{dec}} \frac{dz}{E(z)} $$
In both cases the direct dependence is cancelled out since $H_0 \sim h$. But keeping in mind that $\omega_{\Lambda} = h^2 - \omega_m - \omega_r$ only affects $E(z)$ for small $z$, it will not play a role in the integral for $r_s$, which only goes from $z_{dec} \approx 1090$ upwards. Therefore increasing $h$ will increase $\omega_{Lambda}$ which will increse $E(z)$. That will finally decrease $d_A^c(z_{dec})$ which leads to a stretch of the spectrum as already observed.

** Objective 4: vary $\omega _b$ **

```python
wbvar = np.array([0.01, wb, 0.04])
Cl_val_wbvar = np.empty([len(wbvar), len(l)])
for i in range(len(wbvar)):
    Cl_val_wbvar[i] = np.array(vC(l, wbf=wbvar[i]))
    
for i in range(len(wbvar)):
    plt.plot(l, l*(l+1)*Cl_val_wbvar[i]/(2*math.pi), label=str(wbvar[i]))
plt.legend(); plt.xlabel(xlab); plt.ylabel(ylab)
plt.title(titstr+' for varying $\omega _b$')
plt.savefig('wbvar.pdf'); plt.show(); plt.close()
```

** Objective 5: vary $n_S$ **

```python
nsvar = np.array([0.7, ns, 1.2])
Cl_val_nsvar = np.empty([len(nsvar), len(l)])
for i in range(len(nsvar)):
    Cl_val_nsvar[i] = np.array(vC(l, nsf = nsvar[i]))
    
for i in range(len(nsvar)):
    plt.plot(l, l*(l+1)*Cl_val_nsvar[i]/(2*math.pi), label=str(nsvar[i]))

# multipolo correspondiente a la escala del pivote
k0 = 0.05 # in 1/Mpc
lpiv = kpiv*dAc(zdec, h, wm, wr)
print('l corresponding to $k_0$: l_0=', lpiv)
plt.axvline(x=np.round(lpiv), linestyle=":", label="$l=l_0$")

plt.legend(); plt.xlabel(xlab); plt.ylabel(ylab)
plt.title(titstr+' for varying $n_s$')
plt.savefig('nsvar.pdf'); plt.show(); plt.close()
```
